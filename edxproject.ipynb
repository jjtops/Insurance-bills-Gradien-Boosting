```python
#importing libraries
import pandas as pd
import sklearn
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import Lasso
import seaborn as sns
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Perceptron
```

```python
#removing null values and replaceing categorical data with numbers/standardization has no affect
#on the accuracy of the data
df = pd.read_csv(r'/Users/composerinmaking/Edx Data Science/Edx Project/insurance.csv', sep = ',')
df.dropna
df.sex = pd.Categorical(df.sex, ['male', 'female'],ordered=True)
df.sex = df.sex.cat.codes
df.smoker = pd.Categorical(df.smoker, ['yes', 'no'],ordered=True)
df.smoker = df.smoker.cat.codes
df.region = pd.Categorical(df.region, ['southeast', 'southwest', 'northeast', 'northwest'],ordered=True)
df.region = df.region.cat.codes
df.head()
```

```python
#features dataframe
X = df.drop(columns=['charges'])
X.head()
```

```python
#labels dataframe
y = df['charges']
y.head()
```

```python
#machine learning model/
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
per = Perceptron()
per.fit(X_train, y_train)
#gbrt= GradientBoostingRegressor(n_estimators=100, random_state=0)
#gbrt.fit(X_train, y_train)
print("Training set score: {:.2f}".format(per.score(X_train, y_train)))
print("Test set score: {:.2f}".format(per.score(X_test, y_test)))
```
